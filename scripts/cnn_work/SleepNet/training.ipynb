```python
import torch
import os
from src.cnn.SleepNet.model import SleepNet
from src.cnn.utils.data_loader import make_loaders
from src.cnn.utils.trainer import Trainer
```

```python
# Setup device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device.type)
ROOT_DIR = os.path.split(os.environ['VIRTUAL_ENV'])[0]
```

```python
# Create data loaders
loaders = make_loaders(os.path.join(ROOT_DIR, 'data',"datasets","SleepNet","Dataset"))
test_loader = loaders['test_loader']
val_loader = loaders['val_loader']
train_loader = loaders['train_loader']
```

```python
# Instantiate components
model = SleepNet(p=0.7)
optimizer = torch.optim.RMSprop(model.parameters(), lr=5e-5, weight_decay=5e-5)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
path = os.path.join(ROOT_DIR, r'data\models\SleepNet')

trainer = Trainer(model, optimizer, loss_fn, device, path, use_notebook=True)
```

```python
trainer.fit(train_loader, test_loader, epochs=3)
```

```python
trainer.evaluate(test_loader)
```
